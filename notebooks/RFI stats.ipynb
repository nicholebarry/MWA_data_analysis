```python
import matplotlib.pyplot as plt
import numpy as np

filedir = '/Users/nicholebarry/MWA/data/SSINS_plots/stats/'

fig = plt.figure(figsize=(10,5))
ax2 = fig.add_subplot(2,1,2)
cm = plt.cm.get_cmap('Blues')

for i in range(1,4,1):
    file_data = np.loadtxt(filedir + str(i) + '.txt', usecols=(0,1),skiprows=0)
    percent_unflagged = file_data[:,1]
    valid_inds = np.argwhere(percent_unflagged > 0.4)
    percent_unflagged = percent_unflagged[valid_inds]
    obs_id = file_data[valid_inds,0] - min(file_data[:,0])
    i_arr = np.repeat(i, len(obs_id))
    im = ax2.scatter(obs_id,i_arr,c=percent_unflagged,marker='o',vmin=0,vmax=1,cmap=cm)
    ax2.set_yticks([1,2,3])
    
plt.colorbar(im,ax=ax2)

```
